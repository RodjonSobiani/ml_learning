```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2
h[h < 0] = 0

plt.plot(t,h)
```

```python
g = 9.8
h0 = 100
v0 = 5

t = np.arange(0,5,0.1)

h = h0 - v0 * t - g * t ** 2 / 2
h[h < 0] = 0

plt.plot(t,h)
plt.xlabel('Время, с')
plt.ylabel('Высота, м')
plt.title(f'Свободное падение с высоты {h0} м и начальной скоростью {v0} м/с')
plt.show()
```

```python
g = 9.8
h0 = 50
v0 = 5

t = np.arange(0,2.8,0.1)

h = h0 - v0 * t - g * t ** 2 / 2
h[h < 0] = 0

x = np.arange(0, 4*np.pi, 0.1)
plt.plot(x, np.sin(x), label='Уровень воды')

plt.plot(t,h, label='Падение')
plt.xlabel('Время, с')
plt.ylabel('Высота, м')
plt.title(f'Свободное падение с высоты {h0} м и начальной скоростью {v0} м/с')
plt.legend()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
fig, ax = plt.subplots(1,2, figsize=(16,5))

ax1, ax2 = ax

ax1.plot(np.sin(x))
ax2.scatter(x, 2*np.cos(x))

ax1.set_title('sin')
ax2.set_title('cos')

fig.tight_layout()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
fig, ax = plt.subplots(1,2, figsize=(16,5), sharey=True)

ax1, ax2 = ax

ax1.plot(4*np.sin(x))
ax2.scatter(x, 2*np.cos(x))

ax1.set_title('sin')
ax2.set_title('cos')

fig.tight_layout()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
fig, ax = plt.subplots(2,2, figsize=(16,5), sharey=True, sharex=True)

ax1, ax2 = ax[:,1]

ax1.plot(x, 4*np.sin(x))
ax2.scatter(x, 2*np.cos(x))

ax1.set_title('sin')
ax2.set_title('cos')

fig.tight_layout()
```
