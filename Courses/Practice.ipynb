### Теория вероятности

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
plt.style.use('ggplot')
```

```python
def darts(r1=2, r2=10):
    if r2 < r1:
        return ValueError
    large_circle = plt.Circle((0,0), r2, color='b', alpha=.8)
    small_circle = plt.Circle((0,0), r1, color='r', alpha=.8)
    
    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_xlim(-r2,r2+1)
    ax.set_ylim(-r2,r2+1)
    
    ax.add_artist(large_circle)
    ax.add_artist(small_circle)
    
    ax.set_title('Darts')
    
    result_prob = 1 - (r1/r2) ** 2
    
    print(f'Геометрическая вероятность того, что дротик не попал в красный круг равна: {result_prob}')
    plt.show()
    
darts(r1=2, r2=20)
```

**Распределения:**
- Равномерное дискретное
- Бернулли
- Пуассона

### Линейная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
%matplotlib inline
```

```python
X = np.linspace(-5,5, 20)
X_train = X[0::2].reshape(-1,1)
X_train
```

```python
def generate_data(n_points=20):
    X = np.linspace(-5,5, n_points)
    y = 10 * X - 7
    
    X_train = X[0::2].reshape(-1,1)
    y_train = y[0::2] + np.random.randn(int(n_points/2)) * 10
    
    X_test = X[1::2].reshape(-1,1)
    y_test = y[1::2] + np.random.randn(int(n_points/2)) * 10
    
    print(f'Generated {len(X_train)} train samples and {len(X_test)} test samples')
    
    return X, y, X_train, y_train, X_test, y_test
```

```python
X, y, X_train, y_train, X_test, y_test = generate_data(100)
```

```python
plt.figure(figsize=(10,5))
plt.plot(X,y,label='real')
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(X_train, y_train)

w = model.coef_
b = model.intercept_
```

```python
plt.figure(figsize=(10,5))
plt.plot(X,y,label='real')
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')

plt.plot(X[1::2], X[1::2].reshape(-1,1).dot(w) + b, label='predicted')

plt.legend(loc='best')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_train_predicted = model.predict(X_train)
y_test_predicted = model.predict(X_test)

print('Train MSE: ', mean_squared_error(y_train, y_train_predicted))
print('Test MSE: ', mean_squared_error(y_test, y_test_predicted))
```

```python
from sklearn.metrics import mean_absolute_error

print('Train MAE: ', mean_absolute_error(y_train, y_train_predicted))
print('Test MAE: ', mean_absolute_error(y_test, y_test_predicted))
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
alphas = np.linspace(1, 1000, 100)

weights = np.empty((len(X.columns), 0))


for alpha in alphas:
    ridge_regressor = Ridge(alpha)
    ridge_regressor.fit(X_train, y_train)
    
    weights = np.hstack((weights, ridge_regressor.coef_.reshape(-1,1)))
    
plt.figure(figsize=(15,8))
plt.plot(alphas, weights.T)
plt.show()
```
