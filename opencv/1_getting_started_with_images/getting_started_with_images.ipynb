```python
import os

old_name = r"/content/starry_night.jpg?raw=true"
new_name = r"/content/starry_night.jpg"

os.rename(old_name, new_name)
```

```python
import cv2 as cv
import sys
from google.colab.patches import cv2_imshow

img = cv.imread("starry_night.jpg")
if img is None:
    sys.exit("Could not read the image.")
cv2_imshow(img)
k = cv.waitKey(0)
if k == ord("s"):
    cv.imwrite("starry_night.png", img)
```
